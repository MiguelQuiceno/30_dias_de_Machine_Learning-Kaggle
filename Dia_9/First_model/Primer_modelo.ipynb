```python
#Importar Pandas
import pandas as pd
#Leer los datos
data = pd.read_csv("/home/miguel/Desktop/ML_Notebooks_&_Projects/30_Dias/Dia_9/First_model/melb_data.csv")
data.columns
```

```python
# Ver los datos de arriba
data.head()
```

```python
# Ver datos nulos
data.isnull().sum()
```

```python
#Quitar datos nulos
data = data.dropna(axis = 0)
```

```python
#Ver otra vez datos nulos
data.isnull().sum()
```

```python
# sacar del dataset la variable predictiva
y = data.Price
y
```

```python
# Atributos que usamos para predecir
features = ['Rooms', 'Bathroom', 'Landsize', 'Lattitude', 'Longtitude']
x = data[features]
```

```python
# Descripcion
x.describe()
```

```python
x.head()
```

```python
# Modelo: arboles de desicion
from sklearn.tree import DecisionTreeRegressor
model = DecisionTreeRegressor(random_state = 1)
model.fit(x, y)
```

```python
# Resultados
print("predicciones de estas casas: ")
print(x.head())
print("Precios")
print(model.predict(x.head()))
print("precios actuales")
print(y.head())
```
