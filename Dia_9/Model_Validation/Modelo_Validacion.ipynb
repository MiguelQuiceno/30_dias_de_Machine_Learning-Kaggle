```python
#Importar librerias
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
# Cargar datos y limpiarlos
data = pd.read_csv("/home/miguel/Desktop/ML_Notebooks_&_Projects/30_Dias/Dia_9/First_model/melb_data.csv")
data = data.dropna(axis = 0)
```

```python
# Sacar x y Y
y = data.Price
melbourne_features = ['Rooms', 'Bathroom', 'Landsize', 'BuildingArea', 
                        'YearBuilt', 'Lattitude', 'Longtitude']
X = data[melbourne_features]
X.head()
```

```python
model = DecisionTreeRegressor(random_state = 1)
model.fit(X, y)
```

```python
# El error absoluto promediado
from sklearn.metrics import mean_absolute_error

predicciones = model.predict(X)
print(mean_absolute_error(y, predicciones))
```

### No podemos predecir en el set de entrenamiento

```python
#Conjunto de validacion
from sklearn.model_selection import train_test_split
x_train, x_val, y_train, y_val = train_test_split(X, y, random_state = 1)
New_model = DecisionTreeRegressor()
New_model.fit(x_train, y_train)
```

```python
#Pedicciones
pred = New_model.predict(x_val)
print("Resultados: ", mean_absolute_error(y_val, pred))
```

Hay mucha diferencia en los 2 resultados, la idea es predecir bien en un conjunto en el que modelo no haya sido entrenado
